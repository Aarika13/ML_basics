# 2) histograms

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.randint(18,45,100)
y=np.random.randint(15,400,100)


print(x)
print(y)
```

```python
plt.hist(x)

plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.show()
```

```python
# bins

bins=[15,20,25,30,35,40,45]

plt.hist(x,bins)
plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.show()
```

```python
# rwidth =  bar space 0 to 1


bins=[15,20,25,30,35,40,45]

plt.hist(x,bins,rwidth=0.5)
plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.show()
```

```python
# histtype 


bins=[15,20,25,30,35,40,45]

plt.hist(x,bins,rwidth=0.8,histtype='step')
plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.show()
```

```python
# orientation



bins=[15,20,25,30,35,40,45]

plt.hist(x,bins,rwidth=0.8,orientation='horizontal')
plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.show()
```

```python
# bar color



bins=[15,20,25,30,35,40,45]

plt.hist(x,bins,rwidth=0.8,orientation='horizontal',color='g')
plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.show()
```

```python
# label


bins=[15,20,25,30,35,40,45]

plt.hist(x,bins,rwidth=0.8,orientation='horizontal',color='g',label='age')
plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(x,bins,rwidth=0.8,orientation='vertical',color='g',label='x age')
plt.hist(y,bins,rwidth=0.8,orientation='vertical',color='y',label='y age')


plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist([x,y],bins,rwidth=0.8,orientation='vertical',color=['g','y'],label=['x age','y age'])


plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.legend()
plt.show()
```

```python
# figuresize


plt.figure(figsize=(25,25))


bins=[15,20,25,30,35,40,45]

plt.hist([x,y],bins,rwidth=0.8,orientation='vertical',color=['g','y'],label=['x age','y age'])


plt.title('Students')
plt.xlabel('x age')
plt.ylabel('y label')

plt.legend()
plt.show()
```
