# Handling Missing values

```python
import pandas as pd
df = pd.read_csv('Movieratingdata1.csv')
df
```

# na_values

```python
# change numeric column dtype
df = pd.read_csv('Movieratingdata1.csv', na_values=[462504,'not value'])
df
```

```python
# change numeric column dtype
df = pd.read_csv('Movieratingdata1.csv', na_values={'MOVIE_ID':13905411})
df
```

```python
#here null elements will be true
df.isnull()
# df.notnull()   #opposite to isnull()
```

```python
df.isnull().sum() #total number of null values in each column
```

```python
df.isnull().sum().sum()  # total number of null values
```

# dropna()

```python
import pandas as pd
df = pd.read_csv('Movieratingdata1.csv')
df
```

```python
#if there is any nan value it will not be there
df.dropna()
```

# how:any,all

```python
df.dropna(how='all')
# 'all'--remove the row in which whole rows data is missing
# how='any' works as dropna()
```

# inplace

```python
# removes missing values in specific column
df = pd.read_csv('Movieratingdata1.csv')
df.dropna(inplace=True)
df
```

# pandas fillna method()

```python
#instead of Nan value it will put the value as you enter
# teo types of fillna 1.ffill : for forward fill 2. bfill : for backward fill
df = pd.read_csv('Movieratingdata1.csv')
df.fillna(6)
```

```python
df.fillna({"MOVIE_ID":1000,'MOVIE_TITLE':'Riot2', 'PRODUCTION_YEAR':2050})
```

# join

```python
#different different column names
df1 = pd.DataFrame({'ID1':[1,2,3,4,5],'class':[11,22,33,44,55]})
df2 = pd.DataFrame({'ID2':[1,2,3,4,5], 'Name':[11,22,33,44,55]})
df1.join(df2)
```

# append

```python
df1 = pd.DataFrame({'ID':[1,2,3,4,5],'class':[11,22,33,44,55]})
df2 = pd.DataFrame({'ID':[1,2,3,4,5], 'class':[11,22,33,44,55]})
df1.append(df2)
```
