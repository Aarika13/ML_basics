# Normalization

The main difference betweeen standard scaler and normalization is,
in standard scaler-- the graph can be negative as well as positive(means 1st and 3rd quadrant)
in normalization -- the graph will be always positive(means only in 1st quadrant)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class_label','Alcohol','Malic_Acid']
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Class_label',axis=1),df['Class_label'],random_state=0,test_size=0.20)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train_scale = scaler.transform(X_train) 
X_test_scale = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```
